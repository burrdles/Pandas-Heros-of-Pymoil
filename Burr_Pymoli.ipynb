```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data= pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
#**Count of Total Players**
#Find Unique SN
unique_SN = purchase_data["SN"].unique()
unique_SN
#Count unique SN
total_players = len(unique_SN)
#Display Total Players
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df.head()
```

```python
#Purchase Analysis - Number of Unique Items
unique_items = purchase_data["Item ID"].unique()
total_unique_items = len(unique_items)

#Purchase Analysis - Average Purchase Price
avg_price = round(purchase_data["Price"].mean(),2)

#Purchase Analysis - Total Number of Purchases
unique_purch = purchase_data["Purchase ID"].unique()
total_purchases = len(unique_purch)

#Purchase Analysis - Total Revenue
revenue = round(purchase_data["Price"].sum(),2)
```

```python
#Purchase Analysis - Data Frame
revenue_summary = pd.DataFrame({"Number of Unique Items": [total_unique_items], 
                                "Average Price": [avg_price],
                               "Number of Purchases": [total_purchases],
                               "Total Revenue": [revenue]})

revenue_summary["Average Price"] = revenue_summary["Average Price"].map("${:.2f}".format)
revenue_summary["Total Revenue"] = revenue_summary["Total Revenue"].map("${:,.2f}".format)
revenue_summary
```

```python
#Gender Demographics - new dataframe that removes SN duplicates
SN_df = purchase_data.drop_duplicates(["SN"])
SN_df.head()
```

```python
#Gender Demographics - Summary of Count 
person_count = SN_df["Gender"].value_counts()

#Gender Demographics - Count / Total
gender_perct = round(person_count/total_players,4)*100

#Gender Demographics - Summary Table
gender_count_df = pd.DataFrame({"Total Count": person_count, "Percentage of Players": gender_perct})
gender_count_df
```

```python
# Purchasing Amalysis - Set-up Groupby Gender
gender_group_df = purchase_data.groupby(['Gender'])

# Purchasing Amalysis - how many total purchases by gender 
gender_purch_count = gender_group_df["Gender"].count()

# Purchasing Amalysis - average purchase price by gender 
gender_avg_price = round(gender_group_df["Price"].mean(),2)

# Purchasing Amalysis - total purchase value  by gender 
gender_value = round(gender_group_df["Price"].sum(),2)

# Purchasing Amalysis - average purcahse total per person by gender
average_purch = round(gender_value/person_count,2)
```

```python
# Purchasing Amalysis - new DataFrame
purchasing_analysis = pd.DataFrame({"Purchase Count": gender_purch_count, 
                                   "Average Purchase Price": gender_avg_price, 
                                   "Total Purchase Value": gender_value, 
                                   "Avg Total Purchase per Person": average_purch})

purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis["Total Purchase Value"] = purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis["Avg Total Purchase per Person"] = purchasing_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)

purchasing_analysis
```

```python
# Age Demographics - max age
age_max = purchase_data["Age"].max()
age_max
```

```python
# Age Demographics 
#Create the bins in which Data will be held
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
# Bin Names
age_bins = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Cut Purchase Dataframeinto Bins
purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], bins, labels=age_bins)
purchase_data.head()
```

```python
# Cut Screen Name Dataframe into bins 
SN_df["Age Groups"] = pd.cut(SN_df["Age"], bins, labels=age_bins)
SN_df.head()
```

```python
# Creating a group based off of the bins for Purchase Data
age_group = purchase_data.groupby("Age Groups")
```

```python
# Creating a group based off of the bins for Screen Name / Individal People Data
age_group_SN = SN_df.groupby("Age Groups")
age_group_SN.count()
age_SN_count = age_group_SN["Purchase ID"].count()
age_SN_count
```

```python
# Age Demographics - how many total purchases by age group
age_purch_count = age_group["Purchase ID"].count()
age_purch_count
```

```python
age_SN_perc = round(age_SN_count/(age_SN_count.sum())*100,2)
```

```python
# Age Demographics - new DataFrame
age_demographics_count = pd.DataFrame({"Total Count": age_SN_count, 
                                   "Percentage of Players": age_SN_perc})

age_demographics_count
```

```python
# Age Demographics - average purchase price by gender 
age_price = round(age_group["Price"].mean(),2)

# Age Demographics - average purcahse value by age
total_age_value = round(age_group["Price"].sum(),2)

# Age Demographics - average purcahse per person
average_age_purch = round(total_age_value/age_SN_count,2)
```

```python
# Purchasing Amalysis - new DataFrame
age_demographics = pd.DataFrame({"Purchase Count": age_purch_count, 
                                   "Average Purchase Price": age_price, 
                                   "Total Purchase Value": total_age_value, 
                                   "Avg Total Purchase per Person": average_age_purch})

age_demographics["Average Purchase Price"] = age_demographics["Average Purchase Price"].map("${:.2f}".format)
age_demographics["Total Purchase Value"] = age_demographics["Total Purchase Value"].map("${:,.2f}".format)
age_demographics["Avg Total Purchase per Person"] = age_demographics["Avg Total Purchase per Person"].map("${:.2f}".format)

age_demographics
```

```python
# Top Spenders Amalysis - Groupby Screen Name
SN_group_df = purchase_data.groupby(['SN'])

# Top Spenders Amalysis - Total Purchase Value
SN_purchases = SN_group_df["Price"].sum()

# Top Spenders Amalysis - Average Price
avg_SN = round(SN_group_df["Price"].mean(),2)

# Top Spenders Amalysis - Count of Purchases
count_SN = SN_group_df["Purchase ID"].count()
```

```python
# Purchasing Amalysis - new DataFrame
top_spenders = pd.DataFrame({"Purchase Count": count_SN, 
                             "Average Purchase Price": avg_SN, 
                            "Total Purchase Value": SN_purchases})

top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)

top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:,.2f}".format)


top_spenders.head(5)
```

```python
# Most Popular Items Amalysis - Groupby Item ID
item_group_df = purchase_data.groupby(['Item ID', 'Item Name'])

# Most Popular Items Amalysis - Count of Purchases
item_count = item_group_df["Item ID"].count()

# Most Popular Items Amalysis - Average price
item_price = round(item_group_df["Price"].mean(),2)

# Most Popular Items Amalysis - Total Revenue
item_value = round(item_group_df["Price"].sum(),2)
```

```python
# Most Popular Items - new DataFrame
top_items = pd.DataFrame({"Purchase Count": item_count,
                         "Item Price": item_price,
                         "Total Purchase Value": item_value})

#Most Popular Items - Sorted by Purchase Count
top_item_count = top_items.sort_values("Purchase Count", ascending=False)

top_item_count["Item Price"] = top_item_count["Item Price"].map("${:.2f}".format)
top_item_count["Total Purchase Value"] = top_item_count["Total Purchase Value"].map("${:,.2f}".format)

top_item_count.head(5)
```

```python
#Most Profitable Items - Sorted by Total Venue
top_item_revenue = top_items.sort_values("Total Purchase Value", ascending=False)

top_item_revenue["Item Price"] = top_item_revenue["Item Price"].map("${:.2f}".format)
top_item_revenue["Total Purchase Value"] = top_item_revenue["Total Purchase Value"].map("${:,.2f}".format)

top_item_revenue.head()
```
